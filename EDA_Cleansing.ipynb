```python
tweets3.loc[(tweets3['date'] <  '06/26/19'), 'date_group']='pre-season'  
tweets3.loc[(tweets3['date'] < '10/01/19') & (tweets3['date'] >= '06/26/19'), 'date_group'] = 'na_only' 
tweets3.loc[(tweets3['date'] < '11/11/19') & (tweets3['date'] >= '10/01/19'), 'date_group'] = 'national_season'
tweets3.loc[(tweets3['date'] < '12/18/19') & (tweets3['date'] >= '11/11/19'), 'date_group'] = 'SOE_start'
tweets3.loc[(tweets3['date'] >= '12/18/19'), 'date_group'] = 'rly_hot'
```
